```python
import numpy as np

# Define the Play Tennis dataset
data = [
    ['Sunny', 'Hot', 'High', 'Weak', 'No'],
    ['Sunny', 'Hot', 'High', 'Strong', 'No'],
    ['Overcast', 'Hot', 'High', 'Weak', 'Yes'],
    ['Rain', 'Mild', 'High', 'Weak', 'Yes'],
    ['Rain', 'Cool', 'Normal', 'Weak', 'Yes'],
    ['Rain', 'Cool', 'Normal', 'Strong', 'No'],
    ['Overcast', 'Cool', 'Normal', 'Strong', 'Yes'],
    ['Sunny', 'Mild', 'High', 'Weak', 'No'],
    ['Sunny', 'Cool', 'Normal', 'Weak', 'Yes'],
    ['Rain', 'Mild', 'Normal', 'Weak', 'Yes'],
    ['Sunny', 'Mild', 'Normal', 'Strong', 'Yes'],
    ['Overcast', 'Mild', 'High', 'Strong', 'Yes'],
    ['Overcast', 'Hot', 'Normal', 'Weak', 'Yes'],
    ['Rain', 'Mild', 'High', 'Strong', 'No']
]





label_mapping = {'Yes': 1, 'No': 0}
for i in data:
    # accessing the last element of the instance
    i[-1] = label_mapping[i[-1]]





feature_mapping = {
    'Sunny': 0, 'Overcast': 1, 'Rain': 2,
    'Hot': 0, 'Mild': 1, 'Cool': 2,
    'High': 0, 'Normal': 1,
    'Weak': 0, 'Strong': 1
}
for instance in data:
    for i in range(len(instance) - 1):
        #as last line of each instance is the class so removing that by -1
        instance[i] = feature_mapping[instance[i]]




# Convert data to numpy array
data = np.array(data, dtype=int)



def manhattan_distance(x1, x2):
    return np.sum(np.abs(x1 - x2))






#k = input("Please input the value of K: ")

def k_nearest_neighbors(data, query, k):
    distances = []
    for i in data:
        dist = manhattan_distance(i[:-1], query)
        distances.append((i, dist))
    distances.sort(key=lambda x: x[1])
    # I have a tuple where instances and distance is saved
    #this lambda is a short time function which is
    #make sure that the sorting is being done with the second element by x[1]
    neighbors = [distance[0] for distance in distances[:k]]
    # create list and extract 1st tuple item of first k
    #insatances then put it in the neighbour
    return neighbors


    #distances = [
    #([0, 1, 2, 0], 6.0),
    #([1, 2, 1, 1], 5.5),
    #([0, 1, 0, 1], 8.2),
    #([1, 0, 1, 0], 3.7),
    #([1, 2, 0, 1], 7.1)
#]

def predict_class(neighbors):
    counts = [0, 0] # creating this list where 2nd 0 is for yes and 1st 0 is for no
    for neighbor in neighbors:
        counts[neighbor[-1]] += 1
        #neighbour[-1] is actually accessing the class lebel from the instance
    return 1 if counts[1] > counts[0] else 0



k = int(input("Please input the value of K: "))
query = input("Enter the weather, temperature, humidity, and wind : ")
query_list = query.split()  # Split the input into a list of values

# Map the user-defined query to numerical values using the feature_mapping dictionary
mapped_query = [feature_mapping[value] for value in query_list]
# Query instance
query_instance = np.array(mapped_query, dtype=int)


# Find k-nearest neighbors
neighbors = k_nearest_neighbors(data, query_instance,k)

# Predict the class of the query instance
predicted_class = predict_class(neighbors)
predicted_label = 'Yes' if predicted_class == 1 else 'No'

print(f"Query Instance: {query_instance}")
print(f"Predicted Label: {predicted_label}")
```
